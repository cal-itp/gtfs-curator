# {analysis_name} report

```python
GCS_PATH = (
    "gs://calitp-analytics-data/data-analyses/"
    "gtfs_digest/processed/"
)
analysis_month = "2026-01-01"
file_name = analysis_month.replace("-","_")[0:7]
file_name
```

## operator summary df

```python
import pandas as pd
import gcsfs

operator_df = pd.read_parquet(
    f"{GCS_PATH}fct_monthly_operator_summary_{file_name}.parquet",
    filters=[
        ("Day Type", "==", "Weekday"),
        ("Analysis Name", "==", analysis_name)],
    filesystem = gcsfs.GCSFileSystem()
).drop_duplicates(subset = ["Analysis Name", "Date"])
```

## route gdf

```python
import geopandas as gpd
import google.auth

credentials, project = google.auth.default()

fct_monthly_route_df = gpd.read_parquet(
    f"{GCS_PATH}fct_monthly_routes_{file_name}.parquet",
    filters=[[("Analysis Name", "==", analysis_name)]],
    storage_options={"token": credentials.token}
).reset_index()
```
